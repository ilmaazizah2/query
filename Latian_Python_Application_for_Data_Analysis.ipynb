## Pengelolaan Data `Date`

```python
#7
#Mengubah tipe kolom Date menjadi Datetime
df['order_date']= pd.to_datetime(df['order_date'])
df['ship_date']= pd.to_datetime(df['ship_date'])
df.dtypes
```

```python
#8
#Membuat kolom tambahan untuk day, month, month_num
df['day']=df['order_date'].dt.day_name()
df['month']=df['order_date'].dt.month_name()
df['month_num']=df['order_date'].dt.month
df['year']=df['order_date'].dt.year
df['ship_delay'] = df['ship_date']-df['order_date'] #tipe datetime
df['ship_delay'] = df['ship_delay'].dt.days #konversi ke hari
df.head(5)
```

```python
#9
#Latihan: Filter berdasarkan bulan October tahun 2015
df_oct = df[(df['month_num'] == 10) & (df['year'] == 2015)]
df_oct
```

##Penggunaan fungsi `pivot_table` & `groupby`

### `pivot_table`

**Formula:**

nama_variable  =  `pd.pivot_table`(nama_dataframe`, values=`['kolom_nilai']`,index=`['baris']`,columns=`['header']`,aggfunc=np.`jenis_agregasi)

Output: `DataFrame`

```python
#10
df_pivot = pd.pivot_table(df,values=['quantity'],index=['region'],columns=['category'],aggfunc=np.average)
df_pivot
```

```python
#11
#Buatlah tabel pivot untuk: index = country, header/kolom = city, nilainya = penjumlahan dari profit
df_pivot = pd.pivot_table(df,values=['profit'],index=['country'],columns=['city'],aggfunc=np.sum)
df_pivot
```

### `groupby`

**Formula:**

nama variabel = nama_dataframe`.groupby(by=`['kolom_1'  ,   'kolom_2'])['kolom_agregasi']`.`jenis_agregasi()

Output: `series`


```python
#12
df_groupby = df.groupby(by=["month_num","month"])["profit"].sum()
df_groupby
```

```python
#13
#merubah output groupby menjadi DataFrame dengan as_index = False
df_groupby = df.groupby(by=["city","profit"], as_index=False)["profit"].sum()
df_groupby
```

```python
#14
#Buatlah dengan fungsi groupby, top 5 city penghasil profit paling besar.
df_groupby2 = df.groupby(by=["city"], as_index=False)["profit"].sum().sort_values(by='profit', ascending=False).head(5)
df_groupby2
```

##Membuat dan menggunakan fungsi sederhana pada DataFrame

Formula:

`def` nama_fungsi `(`nilai`):`

    kondisi



```python
#15
#Membuat suatu fungsi
def condition(value):
    if value > 2:
      return 'High'
    elif value == 2:
      return 'Medium'
    else:
        return 'Low'
```

```python
#16
#Pengaplikasian fungsi
df['status'] = df['quantity'].apply(condition)
df.head()
```

```python
#17
#Buatlah suatu kolom baru yang bernama delay_status yang dapat menyatakan bahwa apabila waktu pengiriman <= 4 hari dari tanggal order maka 'Good' selain itu maka 'Bad'
def condition1(value):
    if value <= 4:
      return 'Good'
    else:
        return 'Bad'
df['delay_status'] = df['ship_delay'].apply(condition1)
df.head(5)
```

## Menjalankan SQL di Colab dengan `sqlite3`

```python
#19
#Menjalankan SQL di Colab
from sqlite3 import connect
conn = connect(':memory:')
df.to_sql('tokopaedi',conn, index=False, if_exists='replace')
```

```python
#20
#Query SQL
df_sql = pd.read_sql("""
SELECT
    tokopaedi.city,
    sum(tokopaedi.profit) profit
from
    tokopaedi
group by
    1
order by
    2 desc
limit 10
""", conn)
df_sql
```

```python
#21
#Buat top 5 city berdasarkan sales pada bulan November 2016 dengan menggunakan sql
df_sql1 = pd.read_sql("""
SELECT
  tokopaedi.city,
  sum(tokopaedi.profit) profit
from
    tokopaedi
    where
    tokopaedi.month_num = 11 and
    tokopaedi.year = 2016
group by
    1
order by
    2 desc
limit 5
    """,conn)
df_sql1
```
